```python
from datetime import datetime

def hours_between_dates(date1: str, date2: str) -> float:
    """
    Calculate the number of hours between two dates.
    
    Args:
        date1: First date string in format 'YYYY-MM-DD HH:MM:SS'
        date2: Second date string in format 'YYYY-MM-DD HH:MM:SS'
    
    Returns:
        Number of hours between the two dates (always positive)
    """
    fmt = "%Y-%m-%d %H:%M:%S"
    d1 = datetime.strptime(date1, fmt)
    d2 = datetime.strptime(date2, fmt)
    
    delta = abs(d2 - d1)
    return delta.total_seconds() / 3600

hours_between_dates("2019-01-01 00:00:00", "2023-12-31 23:00:00")
```
