```python
import pandas as pd

# 1. Load your CSV file
# Replace 'electricity_data.csv' with your actual file path
df = pd.read_csv('demand_calendar.csv')

# 2. Define the columns you want to normalize
demand_columns = ['ME', 'NH', 'VT', 'CT', 'RI', 'SEMA', 'WCMA', 'NEMA_BOST']
all_columns = ['ME', 'NH', 'VT', 'CT', 'RI', 'SEMA', 'WCMA', 'NEMA_BOST', 'hour_sin', 'hour_cos', 'dow_sin', 'dow_cos', 'doy_sin', 'doy_cos']

# 3. Apply Min-Max Normalization to scale values between 0 and 1
# Formula: (value - min) / (max - min)
for col in demand_columns:
    col_min = df[col].min()
    col_max = df[col].max()
    df[col] = (df[col] - col_min) / (col_max - col_min)

df[all_columns] = df[all_columns].round(2)

# Alternatively, if you prefer Z-score standardization (mean=0, standard deviation=1), 
# you can use this line instead of the loop above:
# df[demand_columns] = (df[demand_columns] - df[demand_columns].mean()) / df[demand_columns].std()

# 4. Save the normalized dataframe back to a new CSV
# index=False prevents pandas from writing row numbers as a new column
df.to_csv('demand_calendar_normalized.csv', index=False)

print("Normalization complete. File saved as 'demand_calendar_normalized.csv'")
```
